Backpropagation Explained

How it works:
1. Forward Pass: Input data is fed through the network, layer by layer, to produce an output
2. Calculate Loss: The network's output is compared to the actual target output, and a los (error) is calculated.
3. Backward Pass (Backpropagation): The error is propagated backward through the network, from the output layer to the input.
4. Weight Update: The weights and biases are adjusted in the direction that reduces the error, using an optimiztion algorithm.

```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases randomly
        self.W1 = np.random.rand(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.rand(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def forward(self, X):
        # Hidden layer
        self.hidden_input = np.dot(X, self.W1) + self.b1
        self.hidden_output = sigmoid(self.hidden_input)

        # Output layer
        self.output_input = np.dot(self.hidden_output, self.W2) + self.b2
        self.output = sigmoid(self.output_input)
        return self.output

    def backward(self, X, y, output, learning_rate):
        # Calculate error in output layer
        output_error = y - output
        output_delta = output_error * sigmoid_derivative(output)

        # Calculate error in hidden layer
        hidden_error = output_delta.dot(self.W2.T)
        hidden_delta = hidden_error * sigmoid_derivative(self.hidden_output)

        # Update weights and biases
        self.W2 += self.hidden_output.T.dot(output_delta) * learning_rate
        self.b2 += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
        self.W1 += X.T.dot(hidden_delta) * learning_rate
        self.b1 += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output, learning_rate)
            if epoch % (epochs // 10) == 0 or epoch == epochs - 1:
                loss = np.mean(np.square(y - output))
                print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("NeuralNetwork class defined.")
```

```python
# Input data (4 samples, 2 features)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Output data (4 samples, 1 feature) - XOR-like behavior
y = np.array([[0],
              [1],
              [1],
              [0]])

# Network parameters
input_size = 2
hidden_size = 4  # Number of neurons in the hidden layer
o_output_size = 1

epochs = 10000
learning_rate = 0.1

# Create and train the neural network
print("\nTraining the Neural Network...")
nn = NeuralNetwork(input_size, hidden_size, o_output_size)
nn.train(X, y, epochs, learning_rate)

# Test the trained network
print("\nPredictions after training:")
predictions = nn.forward(X)
for i in range(len(X)):
    print(f"Input: {X[i]}, Expected: {y[i]}, Predicted: {predictions[i][0]:.4f}")

# You might notice that the predictions are close to 0 or 1, but not exactly.
# This is typical for sigmoid output, and you'd often apply a threshold (e.g., 0.5) to get binary classification.
```
